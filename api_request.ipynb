```python
from collections import defaultdict

curr_paris =defaultdict(set)

curr_price = {}

for k in resp:
    a,b = k.split('_')
    curr_paris[a].add(b)
    curr_paris[b].add(a)
    curr_price[(a,b)] = float(resp[k]['buy_price'])
    curr_price[(b,a)] = 1/float(resp[k]['buy_price'])

class vertex():
    def __init__(self,name,neighbours):
        self.name = name
        self.neighbours = neighbours
        self.nbrs_nbrs = {}
        self.triangles = set()
    def tell_nbrs_nbrs(self):
        for k in self.neighbours:
            VertexS[k].nbrs_nbrs[self.name]=self.neighbours
    
    def get_my_triangle(self):
        #self.neighbours.remove(self.name)
        for v in self.neighbours:
            common_v = self.neighbours&self.nbrs_nbrs[v]
            for cv in common_v:
                T = tuple(sorted([self.name,v,cv]))
                self.triangles.add(T)
        

VertexS = {}

for k in curr_paris:
    new_v = vertex(k,curr_paris[k])
    VertexS[k] = new_v

for k in VertexS:
    VertexS[k].tell_nbrs_nbrs()

for k in VertexS:
    VertexS[k].get_my_triangle()

tgls = set()
for k in VertexS:
    tgls|=VertexS[k].triangles

###  RESP

earn_dict = {}

fee_log = np.log(0.998**3)

for a,b,c in tgls:
    
    price = [curr_price[(a,b)],curr_price[(b,c)],curr_price[(c,a)]]
    #rever_price = [curr_price[(b,a)],curr_price[(c,b)],curr_price[(a,c)]]
    earn = np.exp(np.log(price).sum()+fee_log)-1
    earn_dict[(a,b,c)] = earn
    earn_dict[(c,b,a)] = np.exp(-np.log(price).sum()+fee_log)-1

sorted(earn_dict.items(),key=lambda x:x[1],reverse=True)
```
